```python
import pandas as pd

df = pd.read_csv(r"C:\Users\ASUS\Desktop\Desktop\Infotact Internship\Healthcare & Medical - Chronic Disease  Prediction and EHR Analysis\dataset\heart_disease_uci.csv")

print("First 5 rows of the dataset:")
display(df.head())
```

```python
print("Dataset information:")
display(df.info())
```

```python
print("Descriptive statistics of the dataset:")
display(df.describe())
```

```python
print('Number of entries with 0 for trestbps:', df[df['trestbps'] == 0].shape[0])
print('Number of entries with 0 for chol:', df[df['chol'] == 0].shape[0])

# Replacing 0 with NaN in 'trestbps' and 'chol'
df['trestbps'] = df['trestbps'].replace(0, float('nan'))
df['chol'] = df['chol'].replace(0, float('nan'))

print('\nAfter replacement:')
print('Number of entries with 0 for trestbps:', df[df['trestbps'] == 0].shape[0])
print('Number of entries with 0 for chol:', df[df['chol'] == 0].shape[0])
```

```python
print("\nUnique values and counts for categorical columns:")
for column in df.select_dtypes(include='object').columns:
    print(f"\n--- Column: {column} ---")
    display(df[column].value_counts(dropna=False))
```

```python
print("\nUpdated Missing Values:")
display(df.isnull().sum())
```

```python
# Convert 'ca' and 'thal' to object type before examining unique values
df['ca'] = df['ca'].astype(str)
df['thal'] = df['thal'].astype(str)

print("\nUnique values and counts for 'ca':")
display(df['ca'].value_counts(dropna=False))

print("\nUnique values and counts for 'thal':")
display(df['thal'].value_counts(dropna=False))
```

```python
# Impute 'ca', 'thal', and 'slope' with 'Missing' category
df['ca'] = df['ca'].fillna('Missing')
df['thal'] = df['thal'].fillna('Missing')
df['slope'] = df['slope'].fillna('Missing')

# Impute 'fbs', 'restecg', 'exang' with mode
for col in ['fbs', 'restecg', 'exang']:
    if df[col].isnull().any():
        mode_val = df[col].mode()[0]
        df[col] = df[col].fillna(mode_val)

# Impute numerical columns with median
for col in ['trestbps', 'chol', 'thalch', 'oldpeak']:
    if df[col].isnull().any():
        median_val = df[col].median()
        df[col] = df[col].fillna(median_val)

print("Missing values after imputation:")
display(df.isnull().sum())
```

```python
print("\nDistribution of the target variable 'num':")
display(df['num'].value_counts(normalize=True))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = ['age', 'trestbps', 'chol', 'thalch', 'oldpeak']

plt.figure(figsize=(15, 12))
for i, col in enumerate(numerical_cols):
    plt.subplot(len(numerical_cols), 2, 2*i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')

    plt.subplot(len(numerical_cols), 2, 2*i + 2)
    sns.boxplot(x='num', y=col, data=df)
    plt.title(f'{col} vs. num')
plt.tight_layout()
plt.show()
```
