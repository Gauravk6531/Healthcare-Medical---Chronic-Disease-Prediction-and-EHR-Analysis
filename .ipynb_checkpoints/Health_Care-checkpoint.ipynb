```python
import pandas as pd

df = pd.read_csv(r"C:\Users\ASUS\Desktop\Desktop\Infotact Internship\Healthcare & Medical - Chronic Disease  Prediction and EHR Analysis\dataset\heart_disease_uci.csv")

print("First 5 rows of the dataset:")
display(df.head())
```

```python
print("Dataset information:")
display(df.info())
```

```python
print("Descriptive statistics of the dataset:")
display(df.describe())
```

```python
# Check unique values and their counts for 'object' type columns
object_cols = df.select_dtypes(include='object').columns
for col in object_cols:
    print(f"\nUnique values for column '{col}':")
    print(df[col].value_counts(dropna=False))
```

```python
# Check values in 'trestbps' and 'chol' that are 0, which might indicate missing values.
print("\nValue counts for 'trestbps' where value is 0:")
print(df[df['trestbps'] == 0]['trestbps'].count())

print("\nValue counts for 'chol' where value is 0:")
print(df[df['chol'] == 0]['chol'].count())
```

```python
# Replace '?' with NaN in relevant columns
df.replace('?', pd.NA, inplace=True)

# Replace 0 with NaN in 'trestbps' and 'chol' as these are not physiologically possible
df['trestbps'] = df['trestbps'].replace(0, pd.NA)
df['chol'] = df['chol'].replace(0, pd.NA)

# Convert 'num' to a binary target variable (0: no disease, 1: disease present)
df['num'] = df['num'].apply(lambda x: 1 if x > 0 else 0)
df.rename(columns={'num': 'heart_disease_present'}, inplace=True)

# Drop the 'id' column as it's not relevant for analysis
df.drop('id', axis=1, inplace=True)

# Display the updated info and head to see changes
display(df.info())
display(df.head())
```

```python
# Display updated info to verify changes, especially NaN counts
display(df.info())

# Display the first few rows to see the new 'heart_disease_present' column and overall structure
display(df.head())
```

```python
# Convert 'trestbps' and 'chol' to numeric, coercing errors to NaN
df['trestbps'] = pd.to_numeric(df['trestbps'], errors='coerce')
df['chol'] = pd.to_numeric(df['chol'], errors='coerce')

# Convert 'fbs' and 'exang' to boolean type
df['fbs'] = df['fbs'].astype('boolean')
df['exang'] = df['exang'].astype('boolean')

# Display updated info to verify data type changes
display(df.info())

# Display the first few rows to see the effect of type conversion
display(df.head())
```

```python
# Impute missing values for numerical columns with the median
for col in ['trestbps', 'chol', 'thalch', 'oldpeak', 'ca']:
    if df[col].isnull().any():
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)

# Impute missing values for categorical columns with the mode
for col in ['fbs', 'restecg', 'exang', 'slope', 'thal']:
    if df[col].isnull().any():
        mode_val = df[col].mode()[0] # .mode() can return multiple values, take the first one
        df[col].fillna(mode_val, inplace=True)

# Display updated info to verify no more missing values
display(df.info())

# Display the first few rows to see the effect of imputation
display(df.head())
```

```python
# One-hot encode categorical columns
categorical_cols = df.select_dtypes(include='object').columns
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Display updated info and head to see changes after one-hot encoding
display(df.info())
display(df.head())
```

```python
# Export the cleaned DataFrame to a CSV file
df.to_csv('heart_disease_cleaned.csv', index=False)

print("Cleaned dataset exported successfully to 'heart_disease_cleaned.csv'")
```
